```python
import pylab, numpy

data = [
    (0.0865, 0.1),
    (0.1015, 0.15),
    (0.1106, 0.2),
    (0.1279, 0.25),
    (0.1892, 0.3),
    (0.2695, 0.35),
    (0.2888, 0.4),
    (0.2425, 0.45),
    (0.3465, 0.5),
    (0.3225, 0.55),
    (0.3764, 0.6),
    (0.4263, 0.65),
    (0.4562, 0.7),
    (0.45, 0.75),
    (0.435, 0.8),
    (0.44, 0.85),
    (0.441, 0.9),
    (0.439, 0.95),
    (0.44, 1.0),
    (0.42, 1.05)
]

def fitData(xs, ys):
    xVals = pylab.array(xs[:]) * 9.81 # get force
    yVals = pylab.array(ys[:])
    pylab.plot(xVals, yVals, 'b*', label="Meausured points")
    
    # Linear Fit
    a,b = pylab.polyfit(xVals, yVals, 1)
    estyVals = a*xVals + b
    print("a =", a, "b =", b)
    pylab.plot(xVals, estyVals, 'r', label="Linear Fit, k = " + str(round(1/a, 5)))
    pylab.legend(loc="best")
    
    # Polynomial Fit
    a,b,c = pylab.polyfit(xVals, yVals, 2)
    estyVals = a*xVals**2 + b*xVals + c
    print("a =", a, "b =", b, "c =", c)
    pylab.plot(xVals, estyVals, 'c', label="Polynomial Fit, k = " + str(round(1/a, 5)))
    pylab.legend(loc="best")
    
    # Display
    pylab.show()
    
def fitData2(xs, ys, degree=1):
    xVals = pylab.array(xs[:]) * 9.81 # get force
    yVals = pylab.array(ys[:])
    
    pylab.figure()
    pylab.plot(xVals, yVals, 'b*', label="Meausured points")
    
    model = pylab.polyfit(xVals, yVals, degree)
    estyVals = pylab.polyval(model, xVals)
    a = model[0]
    k = 1/a
    pylab.plot(xVals, estyVals, "c", label="Best Fit, k = " + str(round(k, 5)))
    pylab.legend(loc="best")
    
    # Display
    pylab.show()
    
    # Goodness of Fit and r Squared
    fit = avgMeanSquareError(yVals, estyVals)
    error = rSquared(yVals, estyVals)
    print("Degree", degree, "| Avg Mean Square Error =", fit, "| R2 =", error)
    
def avgMeanSquareError(data, predicted):
    error = 0
    for i in range(len(data)):
        error += (data[i] - predicted[i])**2
    return error/len(data)

def rSquared(observed, predicted):
    error = ((predicted - observed)**2).sum()
    meanError = error/len(observed)
    return 1 -(meanError/numpy.var(observed))
```

```python
ds = []
ms = []
for d in data:
    ds.append(d[0])
    ms.append(d[1])
    
pylab.plot(ms, ds, "b*")
pylab.show()
```
