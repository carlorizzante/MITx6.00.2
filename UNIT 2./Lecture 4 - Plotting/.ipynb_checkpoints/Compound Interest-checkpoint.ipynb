```python
import pylab
figsize = (16,8)

def retire(monthly, rate, terms):
    """ terms is how many months the retirement plan should be ongoing."""
    savings = [0]
    base = [0] # Basically the months for the plotting, x-axis
    mRate = rate/12
    for i in range(terms):
        base += [i]
        savings += [savings[-1] *  (1 + mRate) + monthly]
    return base, savings

def displayRetirementWMonthlies(monthlies, rate, terms):
    pylab.figure("Retire Month", figsize=figsize)
    pylab.clf()
    for monthly in monthlies:
        xvals, yvals = retire(monthly, rate, terms)
        pylab.plot(xvals, yvals, label="Monthly: " + str(monthly))
        pylab.legend(loc="best")
    pylab.show()
    
def displayRetireWRates(month, rates, terms):
    pylab.figure("Retire Rate", figsize=figsize)
    pylab.clf()
    for rate in rates:
        xvals, yvals = retire(month, rate, terms)
        pylab.plot(xvals, yvals, label="Rate: " + str(month) + " at " + str(int(rate*100)) + "%")
        pylab.legend(loc="best")
    pylab.show()
        
def displayOverview(monthlies, rates, terms):
    pylab.figure("Overview", figsize=figsize)
    pylab.clf()
    pylab.xlim(30*12, 40*12)
    for monthly in monthlies:
        for rate in rates:
            xvals, yvals = retire(monthly, rate, terms)
            pylab.plot(xvals, yvals, label=str(monthly) + " at " + str(int(rate*100)) + "%")
            pylab.legend(loc="best")
    pylab.show()
    
def displayOverviewImproved(monthlies, rates, terms):
    pylab.figure("Overview", figsize=figsize)
    pylab.clf()
    pylab.xlim(30*12, 40*12)
    mvisuals = ["c", "b", "r", "g"]
    rvisuals = ["-", "--", "."]
    for i in range(len(monthlies)):
        for j in range(len(rates)):
            monthly = monthlies[i]
            rate = rates[j]
            visual = mvisuals[i%len(mvisuals)] + rvisuals[j%len(rvisuals)]
            xvals, yvals = retire(monthly, rate, terms)
            pylab.plot(xvals, yvals, visual, label=str(monthly) + " at " + str(int(rate*100)) + "%")
            pylab.legend(loc="best")
    pylab.show()

displayOverviewImproved([x*100 for x in range(5,12,2)], [.03, .05, .07], 40*12)
```

```python
displayRetirementWMonthlies([x*100 for x in range(5,12)], .05, 40*12)
```

```python
displayRetireWRates(800, [.03, .05, .07], 40*12)
```
