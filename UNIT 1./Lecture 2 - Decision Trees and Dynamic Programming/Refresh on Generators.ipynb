```python
def fib(n):
    if n <= 1:
        return n
    else:
        return fib(n-1) + fib(n-2)

def fibGen():
    n1 = 1 # fib(n-1)
    n2 = 0 # fib(n-2)
    yield n2 # returns 0
    yield n1 # returns 1
    while True:
        _next = n1 + n2
        yield _next
        n2 = n1
        n1 = _next
```

```python
fg = fibGen()
for i in range(12):
    print(fib(i), fg.__next__())
```

```python
def fibGen():
    n0 = 0
    n1 = 1
    yield n0
    yield n1
    while True:
        n2 = n0 + n1
        yield n2
        n0 = n1
        n1 = n2
```

```python
f = fibGen()
for i in range(12):
    print(f.__next__())
```

```python
f = fibGen()
count = 0
for i in range(12000):
    n = f.__next__()
    count += 1
print("count:", count, "n:", n)
```
