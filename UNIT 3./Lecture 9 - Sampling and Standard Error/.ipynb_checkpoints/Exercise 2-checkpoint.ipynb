```python
import pylab, numpy

def loadFile():
    file = open('julytemps.txt')
    high = []
    low = []
    for line in file:
        fields = line.split()
        if len(fields) < 3 or not fields[0].isdigit():
            continue
        else:
            high.append(int(fields[1]))
            low.append(int(fields[2]))
    return (low, high)

data = loadFile()
diffTemps = list(numpy.array(data[1]) - numpy.array(data[0]))
pylab.plot(range(1,32), diffTemps)
pylab.show()
```
